# Lab | Not hot songs

```python
import pandas as pd
```

Load the Kaggle dataset 

```python
df = pd.read_csv("/Users/lenovo1/Downloads/Ironhack/dataset.csv")
df.head()
```

Create a new dataframe with the columns that are needed

```python
not_hot_songs = df[['track_name','artists','track_genre']]
not_hot_songs
```

```python
not_hot_songs.isna().value_counts()
```

Drop the rows with null values. There are enough rows, so we can drop those rows without the dataset becoming too small

```python
not_hot_songs.dropna(subset=['track_name'], inplace=True)
not_hot_songs
```

Rename the columns

```python
not_hot_songs.rename(columns={'track_name': 'Song','artists':'Artist'}, inplace=True)
not_hot_songs
```

Load the Billboard Top 100 csv

```python
billboard_top_100 = pd.read_csv("/Users/lenovo1/Downloads/Ironhack/billboard_top.csv")
billboard_top_100.head()
```

Find the songs that are both in the Billboard Top 100 and the Kaggle dataset

```python
already_present_songs = pd.merge(not_hot_songs, billboard_top_100, on=['Song','Artist'])
already_present_songs
```

Drop the rows that contain the songs that are in the Billboard Top 100

```python
not_hot_songs.drop(not_hot_songs[(not_hot_songs['Song'] == 'You Proof') & (not_hot_songs['Artist'] == 'Morgan Wallen')].index, inplace = True)
not_hot_songs.drop(not_hot_songs[(not_hot_songs['Song'] == 'Next Thing You Know') & (not_hot_songs['Artist'] == 'Jordan Davis')].index, inplace = True)
not_hot_songs
```

Select a random sample of 3000 rows

```python
not_hot_songs = not_hot_songs.sample(n=3000)
not_hot_songs
```

Make sure that there are different song genres in the dataframe

```python
not_hot_songs['track_genre'].value_counts()
```

Drop the genre column

```python
not_hot_songs = not_hot_songs[['Song','Artist']]
not_hot_songs
```

Save the dataframe to csv

```python
not_hot_songs.to_csv('not_hot_songs.csv', index=False)
```
